# Rugplot

#### createa basic RUG plot for one Variable

```python
plt.figure(figsize=(15.5,7))
sns.rugplot(data=df,x="Unit price")
```

```python
plt.figure(figsize=(15.5,7))
sns.rugplot(data=df,x="Unit price",y="gross income",)
```

### Group by categorical Variable Using Hue

```python
plt.figure(figsize=(15.5,7))
sns.rugplot(data=df,x="Unit price",y="gross income",hue="Gender")
```

### Combine  This With  A KDF Plot

```python
plt.figure(figsize=(15.5,7))
sns.kdeplot(data=df,x="Unit price",y="gross income",hue="Gender",fill=True)
sns.rugplot(data=df,x="Unit price",y="gross income",hue="Gender")
```

## scatterplot   rugplot


```python
#plt.figure(figsize=(15.5,7))
sns.scatterplot(data=df,x="Unit price",y="gross income",hue="Gender")
sns.rugplot(data=df,x="Unit price",y="gross income",hue="Gender")
```

## Show Rugs Outside OF The Axis

```python
#plt.figure(figsize=(15.5,7))
sns.kdeplot(data=df,x="Unit price",y="gross income",hue="Gender",fill=True)
sns.rugplot(data=df,x="Unit price",y="gross income",hue="Gender",height=-0.04)
```

### Add Histplot

```python
learnerea=sns.FacetGrid(df,col="Outlet_Size")
learnerea.map_dataframe(sns.histplot,x="Sales")
```

### Add Kdeplot

```python
learnerea=sns.FacetGrid(df,col="Outlet_Size")
learnerea.map_dataframe(sns.kdeplot,x="Sales")
```

### Add Lineplot

```python
learnerea=sns.FacetGrid(df,col="Outlet_Size")
learnerea.map_dataframe(sns.lineplot,x="Outlet_Year",y="Sales")
```

```python
learnerea=sns.FacetGrid(df,col="Outlet_Size",row="Outlet_Location_Type")
learnerea.map_dataframe(sns.histplot,x="Sales")
```

```python
learnerea = sns.FacetGrid(df, col="Outlet_Size", row="Outlet_Location_Type", sharey=False, ylim=(0, 1000))
learnerea.map_dataframe(sns.histplot, x="Sales")
learnerea.set_axis_labels("Sales", "count")
plt.show()

```

```python
learnerea = sns.FacetGrid(df, col="Outlet_Size", row="Outlet_Location_Type", sharey=False, ylim=(0, 1000))
learnerea.map_dataframe(sns.histplot, x="Sales")
learnerea.set_axis_labels("Sales", "count")
learnerea.set_titles(col_template='{col_name}',row_template='{row_name}')
plt.show()
```

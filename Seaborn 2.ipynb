##### MEAN Marker in the box plot using showmeans attribute and chenge its style Using meanprops

```python
sns.boxplot(x='Payment', y='Total', hue='Gender', showmeans=True, data=df, meanprops={"marker": "o",
                                                                                  "markerfacecolor": "white",
                                                                                  "markersize": "10",
                                                                                  "markeredgecolor": "black"})
```

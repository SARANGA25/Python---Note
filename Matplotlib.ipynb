```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Company =['Googl','Amzn','MSFT','FB','APPLE']
Revenue =['27','89','90','139']
```

```python
Ypos=np.arange(len(Company))
Ypos
```

```python
Company = ['Googl','Amzn','MSFT','FB','APPLE']
Revenue = [10, 20, 25, 25, 30]
profit = [25, 12, 34, 14, 25]

plt.xticks(Ypos, Company)
plt.ylabel('Revenue (bln)')
plt.title("Tech Stock Revenues")
plt.bar([pos + 0.2 for pos in Ypos], Revenue, width=0.4, label="Revenue")
plt.bar([pos - 0.2 for pos in Ypos],profit, width=0.4, label="profit")
plt.legend()
plt.show()

```

### Horizontal bar

```python
plt.yticks(Ypos, Company)  # Set the y-ticks to display company names
plt.xlabel('Amount (bln)')  # Label for the x-axis
plt.ylabel('Company')  # Label for the y-axis
plt.title("Tech Stock Revenues and Profits")  # Title of the plot
plt.barh([pos + 0.2 for pos in Ypos], Revenue, height=0.4, label="Revenue")  # Plotting revenue data
plt.barh([pos - 0.2 for pos in Ypos], profit, height=0.4, label="profit")  # Plotting profit data
plt.legend()  # Display the legend
plt.show()
```

```python
arr = np.array([[1,5,3],[4,5,9]])
for i in np.nditer(arr):
    print(i)
```
